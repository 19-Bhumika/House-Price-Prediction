# -->Use the train_test_split method available in SCIKIT to split the dataset into Train Dataset and Test Dataset.

# -->Show the Regression Score, Intercept and other parameters etc in the Output.

```python
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
label_encoder=preprocessing.LabelEncoder()
df['BuildingType']=label_encoder.fit_transform(df['BuildingType'])
df['BuildingType'].unique()
df['Location']=label_encoder.fit_transform(df['Location'])
df['Location'].unique()
df['Size']=label_encoder.fit_transform(df['Size'])
df['Size'].unique()

X=df.drop('RentPerMonth',axis=1)
y=df.RentPerMonth
split_data=[]
```

```python
r2_list=[]
r_intercept_list=[]
r_slope=[]
split_ratio=[]
y_p_list=[]
for i in range(50,81,10):
    a,b,c,d=train_test_split(X,y,train_size=i/100,random_state=0)
    split_data.append((a,b,c,d))
    model=LinearRegression().fit(a,c)
    y_predict=model.predict(b)
    y_p_list.append(y_predict)
    r2_list.append(r2_score(d,y_predict))
    r_intercept_list.append(model.intercept_)
    r_slope.append(model.coef_)
    split_ratio.append(f"{i}:{100-i}")

pos=r2_list.index(max(r2_list))

print("The maximum score got is in ratio",split_ratio[pos],"with the accuracy:",max(r2_list))
print("The intercept of the split ratio",split_ratio[pos],"is:",r_intercept_list[pos])
print("The slope of the split ratio",split_ratio[pos],"is:",r_slope[pos])

X_train,X_test,y_train,y_test=split_data[pos] #a,b,c,d for 60% training data
y_pred=y_p_list[pos]
```

# 2.What are the different Error Measures (Evaluation Metrics) in relation to Linear Regression? How much do you get in the above cases?

# 3.Note down the errors/losses when the train-test ratio is 50:50, 60:40, 70:30, and 80:20

## Mean Absolute Error

```python
from sklearn.metrics import mean_absolute_error
for i in range(len(split_data)):
    mae_loss=mean_absolute_error(split_data[i][3], y_p_list[i])
    print("The Mean Absolute Error of",split_ratio[i],"ratio is:",mae_loss)
```

## Mean Squared Error

```python
from sklearn.metrics import mean_squared_error
for i in range(len(split_data)):
    mse_loss=mean_squared_error(split_data[i][3], y_p_list[i])
    print("Mean Squared Error of",split_ratio[i],"ratio is:",mse_loss)
```

## Mean Residual Error

```python
from statistics import mean 
for i in range(len(split_data)):
    residual_error = split_data[i][3]-y_p_list[i]
    mean_r=mean(residual_error)
    print("Mean Residual Error of",split_ratio[i],"ratio is:",mean_r)
```

## Root Mean Squared Error

```python
from math import sqrt
for i in range(len(split_data)):
    rmse_loss=sqrt(mean_squared_error(split_data[i][3], y_p_list[i]))
    print("Root Mean Squared Error of",split_ratio[i],"ratio is:",rmse_loss)
```

# 4.During LinearRegression() process, what is the impact of giving TRUE/FALSE as the value for Normalize Parameter?

```python
for i in range(len(split_data)):
    model=LinearRegression(normalize=True).fit(split_data[i][0],split_data[i][2])
    y_pred=model.predict(split_data[i][1])
    r_score=r2_score(split_data[i][3],y_pred)
    print("\nR_score of ",split_ratio[i]," when normalize is TRUE is: ",r_score)
    
    model=LinearRegression(normalize=False).fit(split_data[i][0],split_data[i][2])
    y_pred=model.predict(split_data[i][1])
    r_score=r2_score(split_data[i][3],y_pred)
    print("R_score of ",split_ratio[i]," when normalize is FALSE is: ",r_score)
```

### Hence, we could observe that normalize can be either True or False, it doesn't make any difference in the accuracy of the model. We could see that with separate split ratios.
